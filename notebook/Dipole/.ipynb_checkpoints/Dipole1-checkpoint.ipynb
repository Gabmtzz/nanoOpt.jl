```julia
function plotImag(ett::Matrix{Float64},NA::Number)
    BB = bbox(0.0,0.75,1.,0.25)
    heatmap(x./1000,y./1000,ett,xlabel=L"x/M~[\mu m]",ylabel=L"y/M~[\mu m]",c=:Greys,colorbar=:none)
    annotate!(1.2,1.6,"NA=$(NA)")
    plot!(x,ett[50,:],c=:black,ls=:dash,bg=:white,lw=1.2,grid=false,xaxis=false,yticks=false,label="",inset=(1,BB),subplot=2)
end
```

```julia
mat1 = material(1.33^2,1.0)
mat2 = material(1.0,1.0)

k₀ = 2π/620.0
NA = 1.0

lens=lensimage(mat1,mat2,k₀,NA)

dip=[0.,0.,1.]
pos=[0.,0.0,0.]

dip=dipole(mat1,k₀,pos,dip);
```

```julia
farf = farfield(dip,lens.dir);
```

```julia
x = 2000*collect(range(-1.,1.,length=101))
y = x = 2000*collect(range(-1.,1.,length=101));
```

```julia
e=efieldC(lens,farf,x,y);
```

```julia
ett = real.(sum(conj(e).*e,dims=3)[:,:,1])
plotImag(ett,NA)
```

```julia
NAa = [0.8,1.0,1.2];
```

## Dipole in x 

```julia
dip=[1.,0.,0.]
dip=dipole(mat1,k₀,pos,dip)
pIMax = []
for NA in NAa
    lens=lensimage(mat1,mat2,k₀,NA)
    farf = farfield(dip,lens.dir)
    e=efieldC(lens,farf,x,y)
    ett = real.(sum(conj(e).*e,dims=3)[:,:,1])
    p = plotImag(ett,NA)
    push!(pIMax,p)
end
```

```julia
dpx = plot(pIMax...,layout=(1,3),size=(800,400))
```

## Dipole in z 

```julia
dip=[0.,0.,1.]
dip=dipole(mat1,k₀,pos,dip)
pIMaz = []
for NA in NAa
    lens=lensimage(mat1,mat2,k₀,NA)
    farf = farfield(dip,lens.dir)
    e=efieldC(lens,farf,x,y)
    ett = real.(sum(conj(e).*e,dims=3)[:,:,1])
    p = plotImag(ett,NA)
    push!(pIMaz,p)
end
```

```julia
dpz = plot(pIMaz...,layout=(1,3),size=(800,400))
```
