```julia
mat = material(1.33^2,1.0)

λ = 620
k₀ = 2π/λ
NA = 1.2
lens = lensfocus(mat,k₀,NA);
```

```julia
w₀ = 0.75*lens.rad
pos = 500*range(-1,1,length=61)
```

```julia
x,y = ndgrid(pos,pos);
```

```julia
z = 0.0
e = efocusHG00(lens,w₀,Matrix(x),Matrix(y),0.0);
```

```julia
Ett  = real.(sum(conj.(e).*e,dims=3)[:,:,1])
heatmap(pos,pos,Ett,xlabel="X",ylabel="Y")
```
